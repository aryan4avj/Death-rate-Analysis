# Death Rate analysis of India and its neighbouring countries.

It is the death-rate data of world from 1960-2018, it is based on death rate per 1,000 people.

Data Link- http://api.worldbank.org/v2/en/indicator/SP.DYN.CDRT.IN?downloadformat=excel

I am analysing data of India and its neighbouring countries, it includes Bangladesh, Bhutan, China, Myanmar, Nepal, Pakistan, Indonesia, Thailand, Sri Lanka, Maldives. i have analysed data from 1960-2018. n
Research Question:-
1. Which Country has the least death rate and state its reason?
2. Which Country has improved their countries death rate fastest and what can be the reason for it?

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
pd.read_excel?
```

```python
df = pd.read_excel('death_rate_excel.xls', sheet_name='Data', skiprows=3)

df = df[(df['Country Name'] == 'India') | (df['Country Name'] == 'Bangladesh') | (df['Country Name'] == 'Bhutan') 
        | (df['Country Name'] == 'China') | (df['Country Name'] == 'Myanmar') | (df['Country Name'] == 'Nepal') 
        | (df['Country Name'] == 'Pakistan') | (df['Country Name'] == 'Indonesia') | (df['Country Name'] == 'Thailand')
        | (df['Country Name'] == 'Sri Lanka') | (df['Country Name'] == 'Maldives') ]

df.drop(["Indicator Name", "Indicator Code", "Country Code"], axis=1, inplace=True)
# df = df[['Country Name', '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018']]
df = df.set_index('Country Name')
df[:] = df[:].apply(pd.to_numeric, errors='coerce')
df = df.T

plt.style.use('seaborn-colorblind')
df.plot(colormap='Set1', figsize=(12,5), lw=4)
plt.xlabel('Years')
plt.ylabel('Death rate, crude (per 1,000 people)')
plt.title('India and Neighboring Countries Death-rate 1960-2018')
plt.show()
```
